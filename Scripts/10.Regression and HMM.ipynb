# Supervised Learning
## Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#bring in the hard drive dataset
df = pd.read_csv(r"C:\Users\sheet\Dataset\data_Q2_2022\dataq2.csv",encoding='utf-8',header=0)
df.head()
```

```python
# Remove rows with missing values
df = df.fillna(0)
```

```python
# number of hdd
print("number of hdd:", df['serial_number'].value_counts().shape) 

# number of different types of harddrives
print("number of different harddrives", df['model'].value_counts().shape)
```

```python
df["end_date"] = df.groupby("serial_number")["date"].transform("max")
```

```python
df.sort_values(by='date').head()
```

```python
df["end_date"] = pd.to_datetime(df["end_date"])
df["date"] = pd.to_datetime(df["date"])
```

```python
df["date_diff"] = df["end_date"] - df["date"]
df["date_diff"].describe()
```

```python
# create a new column 'Unique_value' which is the sum of 'serial_number' and 'model'
df['Unique_value'] = df['serial_number'] + df['model']
```

```python
filter_col = [col for col in df if col.endswith('_raw')]
data = df.drop(columns= filter_col)
```

```python
data.info()
```

```python
# Convert the days column to integer data type
#data['date_diff'] = data['date_diff'].astype(int)
# Convert the timedelta column to an integer column representing the number of seconds
data['date_diff'] = data['date_diff'].apply(lambda x: x.total_seconds()).astype(float)

# Convert the timedelta column to a float column representing the number of seconds
#df['float_seconds'] = df['timedelta'].apply(lambda x: x.total_seconds()).astype(float)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split



# Get the unique combinations of serial_number and model
unique_combinations = data[['serial_number', 'model']].drop_duplicates()

# Split the unique combinations into training and testing sets
train_combinations, test_combinations = train_test_split(unique_combinations, test_size=0.3, random_state=42)

# Filter the original dataset by the training and testing combinations
train = pd.merge(data, train_combinations, on=['serial_number', 'model'])
test = pd.merge(data, test_combinations, on=['serial_number', 'model'])

# Check that each unique combination of serial number and model is not present in both sets
assert len(set(train_combinations.index) & set(test_combinations.index)) == 0
```

```python
train.head()
```

## Feature selection technique 1:Recursive Feature Elimination:

```python
from sklearn.feature_selection import RFE
# Separate the target variable
X = data.drop(['date_diff','failure','date','serial_number','model','capacity_bytes','Unnamed: 0','end_date','Unique_value'], axis=1)
y = data['date_diff']

# Create a linear regression object
model = LinearRegression()

# Create an RFE object and select the number of features to keep
rfe = RFE(model, n_features_to_select=5)

# Fit the RFE object to the data
rfe.fit(X, y)

# Print the selected features
print(X.columns[rfe.support_])
```

```python
# Define the input features and output variable for the regression model
input_cols = ['smart_166_normalized', 'smart_168_normalized', 'smart_174_normalized',
       'smart_231_normalized', 'smart_235_normalized']
output_col = 'date_diff'
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Train the regression model
reg = LinearRegression()
reg.fit(train[input_cols], train[output_col])

# Evaluate the model on the testing data
y_pred = reg.predict(test[input_cols])
mse = mean_squared_error(test[output_col], y_pred)

print(f"Mean squared error: {mse:.2f}")
```

## Feature selection technique 2:MI feature selection 

```python
#for regression problem

# establish target and features of the data
X = data.drop(['date_diff','failure','date','serial_number','model','capacity_bytes','Unnamed: 0','end_date','Unique_value'], axis=1)
y = data['date_diff']
##get train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression

# Select the top K features based on their mutual information with the target variable
k = 10
selector = SelectKBest(score_func=mutual_info_regression, k=k)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Train the regression model
reg = LinearRegression()
reg.fit(X_train_selected, y_train)

# Evaluate the model on the testing data
y_pred = reg.predict(X_test_selected)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean squared error: {mse:.2f}")
```

## Feature selection technique 3:Lasso Regression:

```python
from sklearn.linear_model import LassoCV
import pandas as pd

X = data.drop(['date_diff','failure','date','serial_number','model','capacity_bytes','Unnamed: 0','end_date','Unique_value'], axis=1)
y = data['date_diff']
# Create a LassoCV object
model = LassoCV()

# Fit the model to the data
model.fit(X, y)

# Get the coefficients and select the non-zero ones
coef = pd.Series(model.coef_, index=X.columns)
selected_features = coef[coef != 0].index.tolist()

# Print the selected features
print(selected_features)
```

# Unsupervised Learning
## Hidden Markov Chain

```python
from hmmlearn import hmm
# Train an HMM with two hidden states
model = hmm.GaussianHMM(n_components=2)

# Fit the model to the training data
model.fit(train[input_cols])
```

```python
# Evaluate the model on the testing data
y_pred = model.predict(test[input_cols])
y_pred = np.argmax(y_pred) # convert probabilities to class labels
# Compute the accuracy of the model
accuracy = sum(y_pred == test[output_col]) / len(test[output_col])
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import classification_report
# Print classification report
if len(np.unique(y_pred)) == 1:
    # Handle the case where all predictions are the same
    print("All predictions are the same.")
else:
    y_pred = np.argmax(y_pred, axis=1)
    # Print classification report
    print(classification_report(y_test, y_pred))
```

```python

# Calculate the log-likelihood of the observed data under the model
log_likelihood = model.score(train[input_cols])

print("Log-likelihood:", log_likelihood)
```
